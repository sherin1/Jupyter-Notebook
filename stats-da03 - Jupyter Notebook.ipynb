```python
import numpy as np
```

```python
# Creating a numpy array for stats
a = np.array([[10,20,40],[60,70,80],[30,50,90]])
a
```

```python
#finding minimum of the whole array
np.min(a)
```

```python
#finding maximum of the whole array
np.max(a)
```

```python
#finding minimum of the number in column
np.min(a,axis=0)
```

```python
#finding minimum of the number in row
np.min(a,axis=1)
```

```python
#finding maximum of the number in column
np.max(a,axis=0)
```

```python
#finding maximum of the number in row
np.max(a,axis=1)
```

```python
#finding the average value
np.average(a)
```

```python
#finding the average value for columns
np.average(a,0)
```

```python
#finding the average value for rows
np.average(a,1)
```

```python
#creating a weighted values for the array
w = [2,3,4]
```

```python
#need to provide axis because it cant take 3 elements for 9 elements in array
#column weighted average ((10*2+60*3+30*4)/9,((20*2+70*3+50*4)/9),((40*2+80*3+90*4)/9)
np.average(a,weights=w,axis=0)
```

```python
#need to provide axis because it cant take 3 elements for 9 elements in array
#row weighted average
np.average(a,weights=w,axis=1)
```

# Skewness is the measure of the asymmetry of the distribution of data.

In a Positively skewed distribution, the data values are clustered around the left side of the distribution and the right
side is longer. The mean and median will be greater than the mode in the positive skew.
In a Negatively skewed distribution, the data values are clustered around the right side of the distribution and the left
side is longer. The mean and median will be less than the mode.
